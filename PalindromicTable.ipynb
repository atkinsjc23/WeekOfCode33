Lucy has a table of digits (0-9) with n  rows and m columns. She considers each cell of the table as a rectangle. The top left cell has the coordinates of (0,0) and the bottom right cell has the coordinates of (n-1,m-1). Cells can be grouped together to form larger rectangles.

Lucy is on the lookout for palindromic sequences. She glances at the table and tries to find the sub-rectangle with the largest area such that its digits form a palindromic sequence. She calls such a rectangle a beautiful rectangle. She notices that there can be multiple such rectangles present and is happy to consider any one of them as long as the palindromic sequence formed by its digits don't contain a leading zero. Note that a sequence of only 1 element can contain 0, but 00 is invalid as it contains a leading zero. Can you help Lucy find and print the area and coordinates of the largest beautiful rectangle?

In the first line print the largest area. 
In the second line print the rectangle x)_1, y_1, x_2, y_2 where x_1<=x_2 and y_1<=y_2
If there are multiple beautiful rectangles with the largest area, output any of them.

```python
#!/bin/python3

import sys
def isPal(l,n):
    s=0
    if 0 in l:
        if l[0]>=n-1:
            return(False)
        else:
            for i in l:
                x=l[i]%2
                s+=x
                if s>1:
                    return(False)
            return(True)
    else:
        for i in l:
            x=l[i]%2
            s+=x
            if s>1:
                return(False)
        return(True)

        
def makeLen(x,y):
    d={}
    l=[]
    for i in range (1,x+1):
        for j in range (1,y+1):
            if i*j in d:
                d[i*j].append([i,j])
            else:
                d[i*j]=[[i,j]]
                l.append(i*j)
    return(d,sorted(l,reverse=True))
        
n, m = input().strip().split(' ')
n, m = [int(n), int(m)]
table = []
for table_i in range(n):
    table_t = [int(table_temp) for table_temp in input().strip().split(' ')]
    table.append(table_t)

def doIt(table,n,m):    
    r=makeLen(n,m)
    dic=r[0]
    L=r[1]
    for i in L:
        if i==1:
            return(1,0,0,0,0)
        else:
            p=dic[i]
            for l in p:
                br=l[0]
                bc=l[1]
                d={}
                for a in range (0,br):
                    for b in range (0,bc):
                        if table[a][b] in d:
                            d[table[a][b]]+=1
                        else:
                            d[table[a][b]]=1
                olddic=dict(d)
                for i in range (0,n-br+1):
                    if i!=0:  
                        for k in range (0,bc):
                            olddic[table[i-1][k]]-=1
                            if table[i+br-1][k]  in olddic:
                                olddic[table[i+br-1][k]]+=1
                            else:
                                olddic[table[i+br-1][k]]=1
                        d=dict(olddic)
                    for j in range (0,m-bc+1):
                        if j!=0:
                            for k in range (0,br):
                                d[table[k+i][j-1]]-=1
                                if table[k+i][j+bc-1] in d:
                                    d[table[k+i][j+bc-1]]+=1
                                else:
                                    d[table[k+i][j+bc-1]]=1
                        if isPal(d,br*bc)==True:
                            start=[i,j]
                            end=[i+br-1,j+bc-1]
                            area=br*bc
                            return(area,start[0],start[1],end[0],end[1])
r=doIt(table,n,m)
print(r[0])
print(r[1],r[2],r[3],r[4])



```
